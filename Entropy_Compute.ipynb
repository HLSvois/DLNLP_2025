```python
# Compute English words enrotpy
import nltk
from nltk.corpus import gutenberg, words
all_text = []
for text in gutenberg.fileids():
    all_text+=list(gutenberg.words(text))
print(len(all_text))

english_vocab = set(words.words())

# 过滤非英文单词
filtered_words = [word for word in all_text if word.lower() in english_vocab]
```

```python
print(len(filtered_words))
from collections import Counter
import math
word_counts = Counter(filtered_words)
print(dict(word_counts))
all_f = sum(dict(word_counts).values())
print(all_f)
entropy = 0
for key, value in word_counts.items():
    p = value/all_f
    entropy += -p * math.log2(p)
print(entropy)
```

```python
from collections import Counter
import math
import matplotlib.pyplot as plt


# 按频率排序，取出前 20 个单词
top_20_words = word_counts.most_common(20)

# 分离单词和频率
words, counts = zip(*top_20_words)

# 打印前 20 个单词及其频率
print("Top 20 most frequent words:", top_20_words)

# 绘制柱状图
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 20 Most Frequent Words")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(counts) * 1.1)  # 留出 10% 的空白

plt.show()
```

```python
import nltk
from nltk.corpus import gutenberg
from collections import Counter
import math

# 下载 Gutenberg 语料库（如果尚未下载）
nltk.download('gutenberg')

# 加载 Gutenberg 语料库中的所有单词
all_text = []
for text in gutenberg.fileids():
    all_text += list(gutenberg.words(text))

# 将所有单词拼接成一个字符串，并提取字母（忽略大小写）
all_letters = "".join(all_text).lower()

# 过滤非字母字符（只保留字母）
all_letters = [char for char in all_letters if char.isalpha()]

print("Total letters:", len(all_letters))

# 统计字母频率
letter_counts = Counter(all_letters)

# 打印部分字母频率（避免输出过多）
print("Sample letter counts:", dict(list(letter_counts.items())[:10]))

# 总字母数
total_letters = sum(letter_counts.values())
print("Total filtered letters:", total_letters)

# 计算字母熵
entropy = 0
for char, count in letter_counts.items():
    p = count / total_letters  # 字母的概率
    entropy += -p * math.log2(p)

print("Entropy of the letter distribution:", entropy)
```

```python
import os
import json

# 定义根目录
root_dir = "/Users/lh/DL-nlp2025/wiki_zh"

# 用于存储所有提取的 text
all_texts = []

# 遍历根目录下的所有子文件夹
for subdir, _, files in os.walk(root_dir):
    for file in files:
        # 构造文件的完整路径
        file_path = os.path.join(subdir, file)
        
        # 检查文件是否是文件（而不是文件夹）
        if os.path.isfile(file_path):
            try:
                # 打开文件并逐行读取
                with open(file_path, 'r', encoding='utf-8') as f:
                    for line in f:
                        # 解析每一行的 JSON 对象
                        data = json.loads(line.strip())
                        
                        # 提取 'text' 字段
                        if 'text' in data and isinstance(data['text'], str):
                            all_texts.append(data['text'])
            except Exception as e:
                # 捕获并打印错误信息（如文件格式不正确）
                print(f"Error processing file {file_path}: {e}")

# 打印提取的 text 数量
print(f"Total texts extracted: {len(all_texts)}")

# 打印前 10 个 text（可选）
print("Sample texts:", all_texts[:10])
```

```python
import jieba
from tqdm import tqdm
from collections import Counter
# 假设 all_texts 是之前提取的 text 列表
# 用于存储分词后的结果
all_words = []

# 对每个 text 进行中文分词
for text in tqdm(all_texts, desc="分词进度", mininterval=5):
    # 使用 jieba 进行分词
    words = list(jieba.cut(text))
    for word in words:
        all_words.append(word)

word_counts = Counter(all_words)
```

```python
print(list(word_counts.items())[:10])
import re
filtered_word_counts = {k: v for k, v in word_counts.items() if re.fullmatch(r"[\u4e00-\u9fff]+", k)}

# 打印过滤后的结果
print(list(filtered_word_counts.items())[:10])
```

```python
total_filtered_words = sum(filtered_word_counts.values())
print("Total filtered words:", total_filtered_words)
import math
# 计算过滤后的熵
entropy = 0
for char, count in filtered_word_counts.items():
    p = count / total_filtered_words  # 每个汉字的概率
    entropy += -p * math.log2(p)     # 熵公式

print("Entropy of the filtered word distribution:", entropy)
```

```python
# 将每个汉字词拆分为单个汉字
char_list = []
for word, count in filtered_word_counts.items():
    char_list.extend([char for char in word for _ in range(count)])  # 拆分汉字并重复 count 次

# 计算拆分后的总汉字数
total_chars = len(char_list)
print("Total characters after splitting:", total_chars)

# 统计单个汉字的频率
char_counts = Counter(char_list)

# 计算单个汉字的熵
entropy = 0
for char, count in char_counts.items():
    p = count / total_chars  # 每个汉字的概率
    entropy += -p * math.log2(p)  # 熵公式

print("Entropy of the single-character distribution:", entropy)
```
